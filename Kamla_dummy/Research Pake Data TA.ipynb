# Konversi Jam (object) ke datetime 

```python
df_illegal['jam'] = pd.to_datetime(df_illegal2['jam'], format='%H:%M')
```

# Melakukan penambahan variabel baru hour dan minute dari jam

```python
df_illegal['hour'] = pd.to_datetime(df_illegal2['jam'], format='%H:%M').dt.hour
df_illegal['minute'] = pd.to_datetime(df_illegal2['jam'], format='%H:%M').dt.minute
```

# # Melakukan penambahan variabel baru tahun,bulan,hari,weekday

```python
# create new columns from date column -- Year, Month, Day, Hour, Minutes, DayOfWeek 
df_illegal['Tahun'] = df_illegal['tanggal'].dt.year
df_illegal['Bulan'] = df_illegal['tanggal'].dt.month
df_illegal['Hari'] = df_illegal['tanggal'].dt.day
df_illegal['Weekday'] = df_illegal['tanggal'].dt.dayofweek
df_illegal = df_illegal.sort_values('tanggal')
df_illegal.head()
```

# Melakukan Plot Grafik berdasarkan Latitutde,longitude terhadap nama kapal

```python
# Sebaran plot dari lokasi pelanggar kapal imigran berdasarkan nama kapal

df_illegal = df_illegal.loc[(df_illegal['latitude']>0)]

sns.lmplot('latitude',
           'longitude',
           data=df_illegal[:],
           fit_reg=False, 
           hue="nama kapal", 
           palette='Dark2',
           size=12,
           ci=2,
           scatter_kws={"marker": "D", 
                        "s": 10}) 
ax = plt.gca()
ax.set_title("Sebaran Kapal Pelaku Imigran Illegal")
```

# Melakukan Plot Grafik  rata-rata kejadian imigran illegal per tahun

```python

plt.figure(figsize=(12,6))
sns.barplot(x='Tahun',
            y='Laut Banda',
            data=df_illegal.groupby(['Tahun'])['tempat kejadian'].value_counts().\
                 unstack().reset_index(),
            color='steelblue').\
            set_title("Kasus Imigran Illegal Per Tahunnya Di Laut Banda")
```

# Melakukan Plot Grafik rata-rata kejadian imigran illegal per bulan

```python
# plot homicides sorted by month

fig, ax = plt.subplots(figsize=(14,6))
month_nms = ['January','February','March','April','May','June','July','August'\
             ,'September','October','November','December']    
fig = sns.barplot(x='Bulan',
                  y='Imigran Ilegal',
                  data=df_illegal.groupby(['Tahun','Bulan'])['kategori'].\
                  value_counts().unstack().reset_index(),
                  color='#808080')
ax.set_xticklabels(month_nms)
plt.title("Total Kasus Imigran Illegal-- All Years")

```

Pada data per bulannya, rata-rata kasus relatif stagnan dimana kasus tertinggi bulan april dan terendah july

# Melakukan Plot Grafik rata-rata kejadian imigran illegal per hari

```python
# plot homicide rates vs. day of the week
fig, ax = plt.subplots(figsize=(14,6))
week_days = ['Minggu','Senin','Selasa','Rabu','Kamis','Jumat','Minggu']    
fig = sns.barplot(x='Weekday',
                  y='Imigran Ilegal',
                  data=df_illegal.groupby(['Tahun','Weekday'])['kategori'].\
                       value_counts().unstack().reset_index(),
                  color='steelblue')
ax.set_xticklabels(week_days)
plt.title('Kasus Imigran Illegal Per Harinya -- All Years')

```

# Melakukan Plot Grafik rata-rata kejadian imigran illegal per jam

```python
# use seaborn barplot to plot homicides vs. hour of the day 
fig, ax = plt.subplots(figsize=(14,6))
fig = sns.barplot(x='hour',
                  y='Imigran Ilegal',
                  data=df_illegal.groupby(['Tahun','hour'])['kategori'].\
                       value_counts().unstack().reset_index(),
                  color='steelblue',
                  alpha=.75)
plt.title('Grafik Kejadian Imigran Illegal Per Jam -- All Years')
```
